```python
import matplotlib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature
from cartopy.examples.arrows import sample_data

matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
```

**Main features**

* Object oriented projection definitions.
* Seamless integration with matplotlib.
* Supports point, line, polygon and raster image transformations between projections.
* Simple interfaces to geographic data (e.g. Shapefiles, map tiles, WMS/WMTS, Natural Earth, ...)

Integrates with matplotlib via:
 - **`projection`** argument to Axes/SubAxes - defines the map you want to produce
 - **`transform`** argument to plotting routines - defines the coordinate system of your data

    ax = plt.axes(projection=ccrs.PlateCarree())
    
    ...
    
    ax.imshow(..., transform=ccrs.Orthographic())

    ...

    plt.show()

## Hello world!

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()
ax.set_title(u'Plate Carrée')
plt.show()
```

## Also works when creating subplots...

```python
ax = plt.subplot(1, 2, 1, projection=ccrs.Orthographic())
ax.coastlines()
ax.stock_img()
ax.set_title('Orthographic')

ax = plt.subplot(1, 2, 2, projection=ccrs.Mercator())
ax.coastlines()
ax.stock_img()
ax.set_title('Mercator')

plt.show()
```

## Can use standard matplotlib techniques

`plt.plot()`, `plt.scatter()`, etc.

```python
ax = plt.axes(projection=ccrs.Mollweide())
ax.coastlines()
ax.stock_img()

# Points format (location, latitude, longitude)
points = [('New York', 40.69, -74.04),
          ('Minsk', 53.94, 27.74)]
names, lats, lons = zip(*points)

# Show the locations ...
plt.plot(lons, lats, 'r*', ms=15, transform=ccrs.PlateCarree())

plt.show()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()

# Interpolation is "linear" in the source coordinate system.
# By choosing Geodetic we get a great circle.
plt.plot(lons, lats, lw=2, transform=ccrs.Geodetic())

plt.show()
```

```python
ax = plt.axes(projection=ccrs.InterruptedGoodeHomolosine())
ax.coastlines()
ax.stock_img()

plt.fill([-60, 20, 20, -60], [-20, -20, 60, 60], transform=ccrs.PlateCarree())

plt.show()
```

```python
x, y, u, v, vector_crs = sample_data(shape=(20, 25))
magnitude = (u ** 2 + v ** 2) ** 0.5

ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.contourf(x, y, magnitude, transform=vector_crs)

plt.show()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.pcolormesh(x, y, magnitude, transform=vector_crs)

plt.show()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.quiver(x, y, u, v, magnitude, transform=vector_crs)

plt.show()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.streamplot(x, y, u, v, transform=vector_crs,
               linewidth=2, density=2, color=magnitude)

plt.show()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

print x.min(), x.max()
print y.min(), y.max()
plt.imshow(magnitude, extent=(x.min(), y.min(), x.max(), y.max()), origin='upper', transform=vector_crs)

plt.show()
```

## Easy access to common datasets and standard interfaces

* Natural Earth
* Web map tilings (Google, OpenStreetMap, ...)
* Open Geospatial Consortium (OGC) web services
 * WMS
 * WMTS
* Shuttle radar topography mission
* GSHHS
* shapely geometries

```python
ax = plt.axes(projection=ccrs.OSGB())
ax.coastlines()
plt.show()
```

```python
ax = plt.axes(projection=ccrs.OSGB())
ax.coastlines('10m')
plt.show()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent((0.0, 54.0, -36.0, 10.0))
ax.coastlines()

# Simple pre-canned definition of a Natural Earth dataset,
# complete with styling.
ax.add_feature(cartopy.feature.LAKES)

# Custom definition.
roads = cartopy.feature.NaturalEarthFeature(category='cultural',
                                            name='roads',
                                            scale='10m',
                                            facecolor='none', edgecolor='red',
                                            lw=0.5)
ax.add_feature(roads)

plt.show()
```

**WMS/WMTS integration requires an end-point URL and a layer name.**

```python
ax = plt.axes(projection=ccrs.PlateCarree())
url = 'http://map1c.vis.earthdata.nasa.gov/wmts-geo/wmts.cgi'
layer_name = 'VIIRS_CityLights_2012'
ax.add_wmts(url, layer_name)
ax.set_extent((-15, 25, 35, 60))
plt.plot(-3.5333, 50.7167, 'r*', ms=20)
plt.show()
```
