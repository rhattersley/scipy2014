Creating maps from EPSG codes
=============================

The [European Petroleum Survey Group](http://www.epsg.org/) (EPSG) maintain a [database of definitions](http://www.epsg-registry.org/) for coordinate reference systems and coordinate transforms. These definitions are referenced via simple, numeric codes - for example 27700 corresponds to the [British National Grid](http://www.ordnancesurvey.co.uk/resources/maps-and-geographic-resources/the-national-grid.html) as used by the Ordnance Survey.

Thanks to their brevity and simplicity, EPSG codes are common currency in the world of geographic information systems (GIS). To help increase the interoperability between GIS and Cartopy we need an easy way to obtain the Cartopy projection which corresponds to a particular EPSG code.

I have added code to cartopy (and created a spin-off project called [pyepsg](https://pypi.python.org/pypi/pyepsg)), which makes this translation as simple as:

    projection = ccrs.epsg(code)

Getting ready
-------------

First up we get the standard imports out of the way ... Move along, nothing to see here.

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.ogc_clients as ogc_clients
import matplotlib.pyplot as plt
```

Switzerland, here we come!
--------------------------

The ability to use EPSG codes also allows us to create maps for which the underlying projection has no public API. For example, the Swiss coordinate reference system described by EPSG:21781 uses the "Hotine oblique Mercator azimuth center" projection.

Let's take a look...

```python
# Choose the Swiss projection EPSG:21781
projection = ccrs.epsg(21781)

plt.figure(figsize=(16, 12))
ax = plt.axes(projection=projection)
plt.title('Switzerland')

# Add a WMS layer for extra bling and to show where we are...
ogc_clients._CRS_TO_OGC_SRS[projection] = 'EPSG:21781'
ax.add_wms('http://wms.geo.admin.ch/',
           'ch.swisstopo.pixelkarte-farbe-pk1000.noscale')

# Add small stars over Bern and Zurich.
ax.plot(7.4500, 46.9500, '*', ms=20, transform=ccrs.Geodetic())
ax.plot(8.55, 47.3667, '*', ms=20, transform=ccrs.Geodetic())
ax.set_global()

plt.show()
```
