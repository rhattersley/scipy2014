**Main features**

* Object oriented coordinate system definitions.
* Seamless integration with matplotlib.
* Supports point, line, polygon and raster image transformations between projections.
* Simple interfaces to geographic data (e.g. Shapefiles, map tiles, WMS/WMTS, Natural Earth, ...)

Integrates with matplotlib via:
 - **`projection`** argument to Axes/SubAxes - defines the map you want to produce
 - **`transform`** argument to plotting routines - defines the coordinate system of your data

    ax = plt.axes(projection=ccrs.PlateCarree())
    
    ...
    
    ax.imshow(..., transform=ccrs.Orthographic())

    ...

    plt.show()

## Hello world!

````python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature
from cartopy.examples.arrows import sample_data

matplotlib.rcParams['figure.figsize'] = (10.0, 8.0)
````

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()
plt.title(u'Plate Carrée')
plt.show()
````

## Also works when creating subplots...

````python
ax = plt.subplot(1, 2, 1, projection=ccrs.Orthographic())
ax.coastlines()
ax.stock_img()
ax.set_title('Orthographic')

ax = plt.subplot(1, 2, 2, projection=ccrs.Mercator())
ax.coastlines()
ax.stock_img()
ax.set_title('Mercator')

plt.show()
````

## Can use standard matplotlib techniques

* `plt.plot()`
* `plt.scatter()`
* `plt.contour()`
* etc.

## Points

````python
ax = plt.axes(projection=ccrs.Mollweide())
ax.coastlines()
ax.stock_img()

# Exeter, UK and Austin, TX
lons = [-3.53, -97.75]
lats = [50.72, 30.25]
plt.plot(lons, lats, 'r*', ms=15, transform=ccrs.Geodetic())

plt.show()
````

## Lines

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.stock_img()

# Interpolation is "linear" in the source coordinate system.
# By choosing Geodetic we get a great circle.
plt.plot(lons, lats, lw=2, transform=ccrs.Geodetic())

plt.show()
````

## Polygons

````python
ax = plt.axes(projection=ccrs.InterruptedGoodeHomolosine())
ax.coastlines()
ax.stock_img()

plt.fill([-60, 20, 20, -60], [-20, -20, 60, 60], transform=ccrs.PlateCarree())

plt.fill([120, 290, 290, 120], [-70, -70, -40, -40], transform=ccrs.PlateCarree())

plt.show()
````

## Contour, pcolor, etc.

````python
x, y, u, v, vector_projection = sample_data(shape=(20, 25))
magnitude = (u ** 2 + v ** 2) ** 0.5
````

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.contourf(x, y, magnitude, transform=vector_projection)

plt.show()
````

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.pcolormesh(x, y, magnitude, transform=vector_projection)

plt.show()
````

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.quiver(x, y, u, v, magnitude, transform=vector_projection)

plt.show()
````

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.quiver(x, y, u, v, magnitude, transform=vector_projection,
           regrid_shape=20)

plt.show()
````

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_xlim(-100, 80)
ax.set_ylim(10, 85)

plt.streamplot(x, y, u, v, transform=vector_projection,
               linewidth=2, density=2, color=magnitude)

plt.show()
````

## Images

````python
import os

from cartopy import config

# get the path of the file. It can be found in the repo data directory.
fname = os.path.join(config["repo_data_dir"],
                     'raster', 'sample', 'Miriam.A2012270.2050.2km.jpg'
                     )
img_extent = (-120.67660000000001, -106.32104523100001, 13.2301484511245, 30.766899999999502)
img = plt.imread(fname)
````

````python
ax = plt.axes(projection=ccrs.LambertConformal())
ax.set_extent([-125, -105, 10, 35], ccrs.Geodetic())

ax.imshow(img, origin='upper', extent=img_extent, transform=ccrs.PlateCarree())
ax.coastlines(resolution='50m', color='black', linewidth=2)

plt.show()
````

## Features

* Points, lines and polygons
* Coordinate system
* Styling

Example:
```
cartopy.feature.NaturalEarthFeature(
    category='cultural', name='roads', scale='10m',
    facecolor='none', edgecolor='red', lw=0.5)
```

````python
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent((0.0, 54.0, -36.0, 10.0))
ax.coastlines()

# Simple pre-canned definition of a Natural Earth dataset,
# complete with styling.
ax.add_feature(cartopy.feature.LAKES)

# Custom definition.
roads = cartopy.feature.NaturalEarthFeature(
    category='cultural', name='roads', scale='10m',
    facecolor='none', edgecolor='red', lw=0.5)
ax.add_feature(roads)

plt.show()
````

##Open Geospatial Consortium (OGC) web services

````python
ax = plt.axes(projection=ccrs.PlateCarree())

url = 'http://map1c.vis.earthdata.nasa.gov/wmts-geo/wmts.cgi'
layer_name = 'VIIRS_CityLights_2012'
ax.add_wmts(url, layer_name)

ax.set_extent((-15, 25, 35, 60))
plt.plot(-3.5333, 50.7167, 'r*', ms=20)
plt.show()
````
